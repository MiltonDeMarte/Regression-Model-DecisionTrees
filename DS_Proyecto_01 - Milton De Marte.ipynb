_*FILTRANDO CON METODO *IQR**

```python
def outliers_remove(df,var):
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    q_min = Q1 - (IQR*1.5)
    q_max = Q3 + (IQR*1.5) 
    return df[((df[var] > q_min) & (df[var] < q_max))]
```

```python
dfIQR = outliers_remove(df2,'surface_total')
dfIQR = outliers_remove(dfIQR,'surface_covered')
dfIQR = outliers_remove(dfIQR,'price')
```

1. Selecciona las columnas `rooms`, `bedrooms`, `bathrooms`, `surface_total`, `surface_covered` y `price`.
1. Descarta aquellas instacias con valores faltantes.

*CARGANDO TODO EL DATASET OTRAVEZ*

```python
df= pd.read_csv('DS_Proyecto_01_Datos_properati.csv')
```

*FILTRANDO PROPIEDADES DE `Capital Federal` POR TIPOS = `Casa`, `Departamento`, `PH`*

```python
mask = df.loc[df.property_type == "Departamento"]
mask1 = df.loc[df.property_type == "Casa"]
mask2 = df.loc[df.property_type == "PH"]

masks = [mask,mask1,mask2]
df1 = pd.concat(masks)

df1 = df1.loc[df.l2 == "Capital Federal"]
df1.shape
```

*FILTRANDO INSTANCIAS CON `surface total` MAYOR A 1000 m2*

```python
df1 = df1[df1.surface_total <= 1000]
df1.shape
```

*FILTRANDO INSTANCIAS CON `surface total` MENOR A 15 m2*

```python
df1 = df1[df1.surface_total >= 15]
df1.shape
```

*FILTRANDO INSTANCIAS CON `price` MAYOR A 40000000 DOLARES*

```python
df1= df1[df1.price <=4000000]
df1.shape
```

*SELECCIONANDO LAS COLUMNAS `rooms`, `bedrooms`, `bathrooms`, `surface_total`, `surface_covered` y `price`*

```python
df1= df1[['rooms', 'bedrooms', 'bathrooms', 'surface_total', 'surface_covered', 'price']]
```

*DESCARTANDO INSTANCIAS CON VALORES FALTANTES*

```python
df1.dropna(inplace=True)
```

**Checkpoint:** deberías obtener un dataset con 81019 instacias y 6 columnas.

```python
df1.shape
```

**_MODELOS A UTLIZAR PARA PREDECIR EL PRECIO DE LAS INSTANCIAS: `DecisionTreeRegressor`, `KNeighborsRegressor`.**


**_METRICA DE EVALUACION = `MEAN ABSOLUTE ERROR (MAE)`**

```python
from sklearn.metrics import mean_absolute_error
```

*DECLARANDO VARIABLES PARA `X` E `y`*

```python
X= df1.drop(columns= "price")
y= df1.price
```

*IMPORTANDO `train test split`*

```python
from sklearn.model_selection import train_test_split
```

*CREANDO VARIABLES PARA EL ENTRENAMIENTO Y LA EVALUACION*

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20, random_state=42)
```

**BENCHMARK**

```python
y_train.shape[0]
```

```python
bm_train = np.ones(y_train.shape[0]) * y_train.mean()
bm_test = np.ones(y_test.shape[0]) * y_test.mean()

bm_y_train_pred= bm_train
bm_y_test_pred= bm_test

bm_mae_train= mean_absolute_error(y_train, bm_y_train_pred)
bm_mae_test= mean_absolute_error(y_test, bm_y_test_pred)
print(f"Error Medio Absoluto del Benchmark en Train: ", bm_mae_train)
print(f"Error Medio Absoluto del Benchmark en Test: ", bm_mae_test)

plt.figure(figsize= (20,8))

sns.distplot(y_train - bm_y_train_pred, bins= 15, label= "Train")
sns.distplot(y_test - bm_y_test_pred , bins= 15, label= "Test")
plt.xlabel("\n Errores", fontsize= 20, color= "b")
plt.legend()
```

*METODO CON `DecisionTreeRegressor`*

```python
from sklearn.tree import DecisionTreeRegressor
```

*CREANDO EL ABROL*

```python
tree_regressor= DecisionTreeRegressor(max_depth= 5, random_state= 42)
```

*ENTRENANDO EL ARBOL*

```python
tree_regressor.fit(X_train, y_train)
```

*PREDICIENDO PARA SU EVALUACION*

```python
y_train_pred= tree_regressor.predict(X_train)
y_test_pred= tree_regressor.predict(X_test)
```

*EVALUANDO*

```python
tree_mae_train= mean_absolute_error(y_train, y_train_pred)
tree_mae_test= mean_absolute_error(y_test, y_test_pred)
```

*GRAFICANDO*

```python
print("Error Absoluto Medio de 'TreesRegressor' en Train: ", tree_mae_train)
print("Error Absoluto Medio de 'TreeRegressor' en Test: ", tree_mae_test)

plt.figure(figsize= (10,8))

plt.subplot(1,2,1)

sns.distplot(y_train - y_train_pred, bins= 15, label= "Train")
sns.distplot(y_test - y_test_pred, bins=15 , label= "Test")
plt.xlabel("\n Errores", fontsize= 14)

ax= plt.subplot(1,2,2)
ax.scatter(y_test, y_test_pred, s=2)
lims = [np.min([ax.get_xlim(), ax.get_ylim()]), np.max([ax.get_xlim(), ax.get_ylim()])]
    
ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
plt.xticks(rotation = 45 , fontsize = 10)
plt.xlabel("y_test", fontsize= 14)
plt.ylabel("y_pred", fontsize= 14)
    
plt.tight_layout()
plt.show()
```

# ***Modificando Hiperparametros***

*MODIFICANDO HIPERPARAMETROS PARA `DecisionTreeRegressor`*     (*MODIFICANDO NIVELES DE PROFUNDIDAD*)

```python
lista_mae_train= []
lista_mae_test= []

max_depths= range(1,30,1)

for i in max_depths:
    
    tree_regressor= DecisionTreeRegressor(max_depth= i, random_state=42)
    tree_regressor.fit(X_train, y_train)
    
    y_train_pred= tree_regressor.predict(X_train)
    y_test_pred= tree_regressor.predict(X_test)
    
    mae_train= mean_absolute_error(y_train, y_train_pred)
    mae_test= mean_absolute_error(y_test, y_test_pred)
    
    lista_mae_train.append(mae_train)
    lista_mae_test.append(mae_test)
```

```python
plt.figure(figsize= (20,10))

plt.subplot(1,2,1)
plt.plot(max_depths, lista_mae_train, 'o-', label= "Train")
plt.plot(max_depths, lista_mae_test, 'o-', label= "Test")
plt.legend()
plt.ylabel("\n Error Medio Absoluto", fontsize= 14, color="b")
plt.xlabel("\n Profundidades", fontsize=14, color="b")
plt.title("Numeros de eror al predecir en realacion a más profundiad \n", fontsize= 16, color="g")
plt.show()
```

*MODIFICANDO HIPERPARAMETROS PARA `KNeighborsRegressor`*     (*MODIFICANDO NUMERO DE VECINOS*)

```python
lista_mae_train= []
lista_mae_test=[]

vecinos = [1,1,2,3,5,8,13,21,34,56,90,146]

for i in vecinos:
    
    knn_regressor= KNeighborsRegressor (n_neighbors= i)
    knn_regressor.fit(X_train, y_train)
    
    y_train_pred= knn_regressor.predict(X_train)
    y_test_pred= knn_regressor.predict(X_test)
    
    mae_train= mean_absolute_error(y_train, y_train_pred)
    mae_test= mean_absolute_error(y_test, y_test_pred)
    
    lista_mae_train.append(mae_train)
    lista_mae_test.append(mae_test)
```

```python
plt.figure(figsize= (20,10))

ax= plt.subplot(1,2,1)
plt.plot(vecinos, lista_mae_train, "o-", label= "Train")
plt.plot(vecinos, lista_mae_test, "o-", label= "Test")
plt.xlabel("\n Vecinos", fontsize= 14, color="b")
plt.ylabel("\n Error Medio Absoluto", fontsize= 14, color= "b")
plt.title("Numeros de eror al predecir en realacion a más vecinos \n", fontsize= 16, color="g")
plt.legend()
plt.xlim()

plt.show()
```
